```python
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from IPython.display import SVG
from graphviz import Source
from IPython.display import display
from IPython.display import HTML
style = "<style>svg{width:70% !important; height:70% !important;}</style>"
HTML(style)
```

```python
dogs_n_cats = pd.read_csv('https://stepik.org/media/attachments/course/4852/dogs_n_cats.csv', encoding = "UTF-8")
```

```python
dogs_n_cats = dogs_n_cats.rename(columns={"Длина": "Lenght", "Высота": "Height", "Шерстист": "Woolly", "Гавкает": "Barks", "Лазает по деревьям": "Climbs", "Вид": "Species"})
```

```python
dogs_n_cats.isnull().sum()
```

```python
X = dogs_n_cats.drop(['Species'], axis=1)
y = dogs_n_cats.Species
```

```python
my_clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
my_clf.fit(X, y)
```

```python
tree.plot_tree(my_clf, feature_names=list(X),
               class_names=['Negative', 'Positive'],
               filled=True);
```
