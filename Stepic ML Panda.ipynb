```python
subset_3 = my_stat[(my_stat['V1']>0) & (my_stat['V3']=='A')]
subset_4 = my_stat[(my_stat['V2']!=10) | (my_stat['V4']>=1)]
subset_3, subset_4
```

```python
my_stat0 = my_stat.assign(V5 = my_stat.V1 + my_stat.V4)
my_stat0.head()
```

```python
my_stat1 = my_stat.assign(V6 = np.log(my_stat.V2))
my_stat1.head()
```
